# Is there a general correlation among columns?

```python
interesting_columns = [
   'Total_Alive_Trees',
   'Total_Dead_Trees', 'Total_Creatures', 'Desert_Water',
   'MountainValley_Water', 'Plains_Water', 'Waterfall_Water',
   'Floor_Water', 'Jungle_Water', 'Reservoir_Water', 'Wetlands_Water',
   'Desert_Clouds', 'MountainValley_Clouds', 'Plains_Clouds',
   'Waterfall_Clouds', 'Jungle_Clouds', 'Reservoir_Clouds',
   'Wetlands_Clouds', 'Desert_Creatures', 'MountainValley_Creatures',
   'Plains_Creatures', 'Jungle_Creatures', 'Reservoir_Creatures',
   'Wetlands_Creatures', 'Desert_Plants', 'Plains_Plants', 'Jungle_Plants',
   'Wetlands_Plants', 'Desert_Users', 'MountainValley_Users',
   'Plains_Users', 'Jungle_Users', 'Reservoir_Users', 'Wetlands_Users']
```

```python
result = []
for i, col in enumerate(interesting_columns):
    
    inner_res = []
    
    for j, col2 in enumerate(interesting_columns):

        corr = data[col].corr(data[col2])
        if not np.isnan(corr):
            if np.abs(corr) > 0.4:
                inner_res.append(corr)
                continue
        inner_res.append(0)
    
    result.append(inner_res)
```

```python
fig, ax = plt.subplots(1,1,figsize=(25,25))
sns.heatmap(result, ax=ax)
ax.set_xticklabels(interesting_columns, rotation='vertical')
ax.set_yticklabels(interesting_columns, rotation='horizontal')
plt.show()
```
