# Look at water

```python
water_subset = data[data.columns[data.columns.str.contains('Water') & (~data.columns.str.contains('Bins|Waterfall'))]]

fig, (ax1, ax2) = plt.subplots(2,1, figsize=(15,8))

water_subset.plot(ax=ax1, legend=None)
ax1.set_ylim([-0.5, 7])
ax1.set_ylabel('Water Amount')
ax1.set_xlabel('Time Step')

water_subset[water_subset.columns[~water_subset.columns.isin(['Total_Water'])]].plot(ax=ax2)
ax1.set_ylabel('Water Amount')
ax1.set_xlabel('Time Step')

plt.show()
```

# Look at Plants

```python
plants_subset = data[data.columns[data.columns.str.contains('lv') & (~data.columns.str.contains('Dead'))]]

fig, axes = plt.subplots(2,2, figsize=(15,8))
axes = axes.flatten()

plants_subset[plants_subset.columns[plants_subset.columns.str.contains('lv1')]].plot(ax=axes[0])
axes[0].set_title('Level 1')

plants_subset[plants_subset.columns[plants_subset.columns.str.contains('lv2')]].plot(ax=axes[1])
axes[1].set_title('Level 2')

plants_subset[plants_subset.columns[plants_subset.columns.str.contains('lv3')]].plot(ax=axes[2])
axes[2].set_title('Level 3')

plants_subset[plants_subset.columns[plants_subset.columns.str.contains('lv4')]].plot(ax=axes[3])
axes[3].set_title('Level 4')

plt.show()
```

```python
plants_subset_dead = data[data.columns[data.columns.str.contains('Dead')]]

fig, axes = plt.subplots(2,2, figsize=(15,8))
axes = axes.flatten()

plants_subset_dead[plants_subset_dead.columns[plants_subset_dead.columns.str.contains('lv1')]].plot(ax=axes[0])
axes[0].set_title('Level 1')

plants_subset_dead[plants_subset_dead.columns[plants_subset_dead.columns.str.contains('lv2')]].plot(ax=axes[1])
axes[1].set_title('Level 2')

plants_subset_dead[plants_subset_dead.columns[plants_subset_dead.columns.str.contains('lv3')]].plot(ax=axes[2])
axes[2].set_title('Level 3')

plants_subset_dead[plants_subset_dead.columns[plants_subset_dead.columns.str.contains('lv4')]].plot(ax=axes[3])
axes[3].set_title('Level 4')

plt.show()
```
